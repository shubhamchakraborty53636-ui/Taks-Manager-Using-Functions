## Project: Simple Task Manager


### Task Manager Functions

We'll define a few functions to handle different aspects of our task manager:

1.  `add_task(task_list, description)`: Adds a new task to the list.
2.  `mark_task_complete(task_list, task_id)`: Marks a task as complete.
3.  `display_tasks(task_list)`: Shows all tasks and their status.
4.  `get_task_by_id(task_list, task_id)`: Helper function to retrieve a task.

```python
# Initialize an empty list to store our tasks
tasks = []

def add_task(task_list, description):
    """Adds a new task to the task list.

    Each task is a dictionary with 'id', 'description', and 'completed' status.
    """
    task_id = len(task_list) + 1
    new_task = {
        "id": task_id,
        "description": description,
        "completed": False
    }
    task_list.append(new_task)
    print(f"Task '{description}' (ID: {task_id}) added.")

def get_task_by_id(task_list, task_id):
    """Helper function to find a task by its ID.

    Returns the task dictionary if found, otherwise None.
    """
    for task in task_list:
        if task['id'] == task_id:
            return task
    return None

def mark_task_complete(task_list, task_id):
    """Marks a task as completed given its ID.
    """
    task = get_task_by_id(task_list, task_id)
    if task:
        if not task['completed']:
            task['completed'] = True
            print(f"Task '{task['description']}' (ID: {task_id}) marked as complete.")
        else:
            print(f"Task '{task['description']}' (ID: {task_id}) was already complete.")
    else:
        print(f"Error: Task with ID {task_id} not found.")

def display_tasks(task_list):
    """Displays all tasks and their completion status.
    """
    print("\n--- Your Tasks ---")
    if not task_list:
        print("No tasks yet. Add some!")
        return
    for task in task_list:
        status = "[X]" if task['completed'] else "[ ]"
        print(f"{status} ID: {task['id']} - {task['description']}")
    print("------------------")
```

```python
# Adding some tasks
add_task(tasks, "Learn Python functions")
add_task(tasks, "Complete the Colab assignment")
add_task(tasks, "Read a book")

# Display tasks to see the current list
display_tasks(tasks)

# Mark a task as complete
mark_task_complete(tasks, 1) # Mark 'Learn Python functions' as complete

# Try to mark a non-existent task
mark_task_complete(tasks, 99)

# Display tasks again to see the updated status
display_tasks(tasks)

# Add another task
add_task(tasks, "Plan weekend activities")
display_tasks(tasks)

# Mark another task complete
mark_task_complete(tasks, 3)
display_tasks(tasks)
```
